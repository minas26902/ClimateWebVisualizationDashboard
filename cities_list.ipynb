```python
#Import dependencies
import pandas as pd
import csv
```

```python
#Import csv file and convert into df
cities=pd.read_csv("resources/city_list_df.csv").drop(['Unnamed: 0'], axis=1)
cities_df = pd.DataFrame(cities)
cities_df['City_Country'] = cities_df['City_Country'].str.upper()
cities_df.head()
```

```python
cities_df[['City_Country', 'Country']] = cities_df['City_Country'].str.split(',', expand=True)
cities_df=cities_df.rename(columns = {'City_Country':'City'})
cities_df=cities_df[['City', 'Country', 'Temperature (F)','Humidity (%)','Cloudiness (%)', 'Wind Speed (mph)', 'Lat', 'Lng']]
cities_df.head()
```

```python
#Convert dataframe to html
cities_df.to_html(open('cities_table.html', 'w'))
#Convert to csv file
cities_df.to_csv("cities_table.csv")
```
